```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("/Users/mjhans/dev/gitRepo/basic_mathmatics/CollaborativeFiltering/CF_data.csv")

df.tail()
```

```python
df.login_id = df.login_id.fillna('None')
df['pl_id'] = df['pcid'] + "_" + df['login_id']
pre_data = df.groupby(['pl_id','product']).count()['action'].unstack(level=1, fill_value=0)
pre_data.tail()
```

```python
def Euclidean_distance(p, q):
	sum = 0
	for idx in range(len(p)):
		sum += math.pow((q[idx]-p[idx]),2)
	return math.sqrt(sum)


def Taxicob_distance(p, q):
	sum = 0
	for idx in range(len(p)):
		sum += abs(p[idx] - q[idx])
	return sum
	
```

```python
apples = np.array(pre_data["apple"]) # p
melons = np.array(pre_data["melon"]) # q
shoes = np.array(pre_data["shoes"]) # q


# 유클리디언 거리
print("apple-melon Euclidean distance: {}".format(Euclidean_distance(apples, melons)))
print("melon-shoes Euclidean distance: {}".format(Euclidean_distance(melons, shoes)))
print("apple-shoes Euclidean distance: {}".format(Euclidean_distance(apples, shoes)))

# Taxicob 거리
print("apple-melon Taxicob distance: {}".format(Taxicob_distance(apples, melons)))
print("melon-shoes Taxicob distance: {}".format(Taxicob_distance(melons, shoes)))
print("apple-shoes Taxicob distance: {}".format(Taxicob_distance(apples, shoes)))
```
